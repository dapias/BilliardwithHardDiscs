#Análisis de la distribución de energías de escape en una celda del sistema tipo billar-disco duro

```julia
using HDF5
using StatsBase
using PyPlot
```

```julia
file = h5open("HDF5/10millones.hdf5")
edata = read(file["edata/"]);
```

```julia
plt.hist(edata,100, normed=1);
```

```julia
ECDF = ecdf(edata) #Empirical Cumulative Distribution Function
```

```julia
intervalo = [0.:0.001:1.0;]; #Energía máxima = 1.0;
```

```julia
CCDF = 1. - ECDF(intervalo)  #Complementary Cumulatice Distribution function;
```

```julia
plot(intervalo, CCDF, ".--")
plot(intervalo, 1. - intervalo, "r+")
```

```julia
plot(intervalo, CCDF, ".--")
plt.gca()[:set_yscale]("log")
plt.gca()[:set_xscale]("log")
```
