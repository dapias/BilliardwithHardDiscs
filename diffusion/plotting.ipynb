```julia
filename = "diezmillones";
```

```julia
file = h5open("HDF5/$filename.hdf5","r")
```

```julia
vnewdisk = read(attrs(file)["vnewdisk"])
```

```julia
t = gettime(filename);
```

```julia
v = getmsdperensemble(filename);
```

```julia
#plot(log(t[400:end]),log(v[400:end]), ".")
#plt.gca()[:set_ylabel](L"$\log(\langle(\Delta x)^2\rangle_t)$")
#plt.gca()[:set_xlabel](L"$\log(t)$")
```

```julia
Rs, slope, intercept = fitmsdwithlinearsquares(log(t[400:end]), log(v[400:end]))
```

```julia
R = string(Rs)
s = string(slope);
```

```julia
fig = plt.figure()
ax = fig[:add_axes]([0.1,0.1,0.9,0.9])
ax[:set_xlabel]("t")
ax[:set_ylabel](L"$\langle(\Delta x)^2\rangle_t$")
ax[:plot](t,v)

logt = log(t[400:end])
logv = log(v[400:end])

ax_inset=fig[:add_axes]([0.55,0.18,0.4,0.33])
ax_inset[:set_ylabel](L"$\log(\langle(\Delta x)^2\rangle_t)$")
ax_inset[:set_xlabel](L"$\log(t)$")
ax_inset[:plot](logt,logv,"b-", label ="data")
ax_inset[:plot](logt, slope*logt + intercept, "r-", label="fit")
corr_text = ax_inset[:text](0.02,0.78,"",transform=ax_inset[:transAxes])
corr_text[:set_text]("\$R^2\$ = $(R[1:6])  \n slope = $(s[1:6])")
handles, labels = ax_inset[:get_legend_handles_labels]()
ax_inset[:legend](handles, labels, loc =4)
fig[:savefig]("./images/cienmildediezmillones/fitasinset.pdf")
```
